Import library pandas,numpy, sklearn dan membaca data csv

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
data = pd.read_csv('player_stats.csv')
```

memcopy datanya dan merubah kolom acs yg object menjadi integer

```python
data_copy = data.copy()
data_copy['acs'] = pd.to_numeric(data_copy['acs'], errors='coerce', downcast='integer')

```

membuat dataset menjadi training set dan testing set dengan proporsi 70:30

di training set dan testing set untuk memprediksi memang kalahnya suatu tim

```python
X = data_copy.drop(columns=['win_lose'])
y = data_copy['win_lose']
proporsi_testing = 0.3

X_training, X_testing, y_training, y_testing = train_test_split(X, y, test_size=proporsi_testing, random_state=42)
X_training.head()

```

 normalisasi data pada attribute acs menggunakan Min Max scaler 

```python
from sklearn.preprocessing import MinMaxScaler

kolom_normalisasi = ['score_team']


scaler = MinMaxScaler()

data_copy[kolom_normalisasi] = scaler.fit_transform(data_copy[kolom_normalisasi])
data_copy
```

Melakukan  standarisasi pada kolom acs,kill,death,assist menggunakan sklearn

```python
from sklearn.preprocessing import StandardScaler
data_copy['acs'] = pd.to_numeric(data['acs'], errors='coerce', downcast='integer')
kolom_standarisasi = [ 'acs', 'kill', 'death', 'assist']
scaler = StandardScaler()
data_copy[kolom_standarisasi] = scaler.fit_transform(data_copy[kolom_standarisasi])

```

Lakukan Data cleaning pada data dengan nilai null dan Ganti nilai null sesuai ketentuan

Menampilkan data yg ada nullnya

```python
baris_dengan_null = data_copy[data_copy.isnull().any(axis=1)]
baris_dengan_null
```

melakukan pergantian data null sesuai tipe datanya

```python
nilai_null = data_copy.isnull().sum()
kolom_dengan_data_null = data_copy.columns[data_copy.any()].tolist()
for kolom in kolom_dengan_data_null:
    if nilai_null[kolom] > 0:
        tipe_data_kolom = data_copy[kolom].dtype
        print(tipe_data_kolom)
        if tipe_data_kolom == 'int64':
            median = data_copy[kolom].median()
            data_copy[kolom].fillna(median, inplace=True)
        elif tipe_data_kolom == 'float64':
            mean = data_copy[kolom].mean()
            data_copy[kolom].fillna(mean, inplace=True)
        elif tipe_data_kolom == 'object':
            modus = data_copy[kolom].mode()[0]
            data_copy[kolom].fillna(modus, inplace=True)
```

Menampilkan lagi data yg sudah diubah kalo kosong berarti datanya tidak ada

```python
baris_dengan_null = data_copy[data_copy.isnull().any(axis=1)]
baris_dengan_null
```

Melakukan Data cleaning pada data dengan nilai duplikat

Mengechek data ada yg duplikat atau tidak

```python
data_duplikat = data_copy[data_copy.duplicated(keep=False)]
data_duplikat
```

Ganti tipe data salah satu attribute angka

```python
data_copy['acs'] = data_copy['kill'].astype(float)
```

melakukan one hot encoding

di dalam kodingan tersebut mengencoding data di kolom win_lose lalu menambahkan kolom baru kalo dengan isi kalo opponent win = 1 sedangkan team win = 0

```python
from sklearn.preprocessing import OneHotEncoder
OneHotEncoder = OneHotEncoder(sparse_output=False)
Win_lose_Encoding = OneHotEncoder.fit_transform(data_copy[["win_lose"]])
Win_lose_Encoding = pd.DataFrame(Win_lose_Encoding)
data_copy['opponent win'] = Win_lose_Encoding[0]
data_copy['team win'] = Win_lose_Encoding[1]
data_copy.head()
```
