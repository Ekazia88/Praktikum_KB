```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


data = pd.read_csv('player_stats.csv')
data['acs'] = pd.to_numeric(data['acs'], errors='coerce', downcast='integer')
```

```python
X = data.drop(columns=['win_lose'])
y = data['win_lose']
proporsi_testing = 0.3

X_training, X_testing, y_training, y_testing = train_test_split(X, y, test_size=proporsi_testing, random_state=42)
```

```python
data_copy = data.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler

kolom_normalisasi = ['acs']


scaler = MinMaxScaler()

data_copy[kolom_normalisasi] = scaler.fit_transform(data_copy[kolom_normalisasi])
data_copy
```

```python
from sklearn.preprocessing import StandardScaler
data_copy['acs'] = pd.to_numeric(data['acs'], errors='coerce', downcast='integer')
kolom_standarisasi = [ 'acs', 'kill', 'death', 'assist']
scaler = StandardScaler()
data_copy[kolom_standarisasi] = scaler.fit_transform(data_copy[kolom_standarisasi])
data_copy
```

```python
data_null = data_copy.isnull()
kolom_dengan_data_null = data_null.columns[data_null.any()].tolist()
for kolom in kolom_dengan_data_null:
    tipe_data_kolom = data[kolom].dtype
    if tipe_data_kolom == 'int64':
        median = data[kolom].median()
        data[kolom].fillna(median, inplace=True)
    elif tipe_data_kolom == 'float64':
        mean = data[kolom].mean()
        data[kolom].fillna(mean, inplace=True)
    elif tipe_data_kolom == 'object':
        modus = data[kolom].mode()[0]
        data[kolom].fillna(modus, inplace=True)
```

```python
data_duplikat = data[data.duplicated(keep=False)]
data_duplikat
```
